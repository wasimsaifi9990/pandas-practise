```python
def grouper(df,idx,col):
    if 80<=df[col].loc[idx]<=90:
        return '80-90'
    elif 50<=df[col].loc[idx]<=60:
        return '50*60'
    else:
        return 'other'
```

```python
g=df.groupby(lambda x:grouper(df,x,'temperature'))
g
```

```python
for key,data in g:
    print('Group by keys: {}'.format(key))
    print('data',data)
```

# <font color="purple"><h3 align="center">Pandas Merge Tutorial</h3></font> 

** DataFrame.merge(right, how='inner', on=None, left_on=None, right_on=None, left_index=False, right_index=False, sort=False, suffixes=('_x', '_y'), copy=True, indicator=False, validate=None) ** 

## parameter
- how{left, right, outer, inner, cross}, default inner 
- on:label or list
- left_on:label or list, or array-like
- right_on:label or list, or array-like

## <font color='blue'>Basic Merge Using a Dataframe Column</font> 

```python
import pandas as pd
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df2
```

```python
df=pd.merge(df1,df2,on='city')
df
```

## <font color='blue'>Type Of DataBase Joins</font> 

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35, 38],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "humidity": [65,68,71],
})
df2
```

```python
df3 = pd.merge(df1,df2,on='city',how='inner')
df3
```

```python
df3 = pd.merge(df1,df2,on='city',how='outer')
df3
```

```python
df3 = pd.merge(df1,df2,on='city',how='left')
df3
```

```python
df3 = pd.merge(df1,df2,on='city',how='right')
df3
```

## <font color='blue'>indicator flag</font> 

```python
df3 = pd.merge(df1,df2,on='city',how='outer',indicator=True)
df3
```

## <font color='blue'>suffixes</font> 

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35,38],
    "humidity": [65,68,71, 75]
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "temperature": [21,14,35],
    "humidity": [65,68,71]
})
df2
```

```python
df3 = pd.merge(df1,df2,indicator=True,suffixes=['_left','_right'],on='city',how='outer')
df3
```

<h2 style="color:purple">Normalize</h2> 

## normalize: bool, {all, index, columns}, or {0,1}, default False
#Normalize by dividing all values by the sum of values.

- If passed all or True, will normalize over all values.
- If passed index will normalize over each row.
- If passed columns will normalize over each column.
- If margins is True, will also normalize margin values. 

```python
pd.crosstab(df.Sex,df.Handedness,normalize='index',margins=True)
```

```python
pd.crosstab(df.Sex,df.Handedness,normalize=True,margins=True)
```
